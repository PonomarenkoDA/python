### Задача №1

Вашему другу, который тоже начал изучать Python, преподаватель дал такую задачу: есть три списка — основной и два побочных. В основном лежат элементы [1, 5, 3], а в побочных — [1, 5, 1, 5] и [1, 3, 1, 5, 3, 3] соответственно. <br/>
Первый побочный закидывается в основной, там считается количество цифр 5, количество выводится на экран, и затем они удаляются из основного списка. После этого в основной закидывается второй побочный список, там считается количество цифр 3 и выводится на экран. В конце также выводится и сам список.<br/>
Из интереса вы попросили вашего друга показать код его программы и поняли, что сделали это не зря — то, что вы увидели, повергло вас в шок и ужас. Вот сам код:

```python
# Условие задачи

a = [1, 5, 3]
b = [1, 5, 1, 5]
c = [1, 3, 1, 5, 3, 3]
for i in b:
    a.append(i)
t = 0
for i in a:
    if i == 5:
        t += 1
print(t)
d = []
for i in a:
    if i != 5:
        d.append(i)
for i in c:
    d.append(i)
t = 0
for i in d:
    if i == 3:
        t += 1
print(t)
print(d)
```

Используя знания о методах списков, а также о стиле программирования, помогите другу переписать программу. Не используйте дополнительные списки.<br/>
Результат работы программы:<br/>
Кол-во цифр 5 при первом объединении: 3<br/>
Кол-во цифр 3 при втором объединении: 4<br/>
Итоговый список: [1, 3, 1, 1, 1, 3, 1, 5, 3, 3]

```python
a = [1, 5, 3]
b = [1, 5, 1, 5]
c = [1, 3, 1, 5, 3, 3]

# Объединение основного и побочного списка
a.extend(b)

# Подсчет количества пятерок
t = a.count(5)
print('Кол-во цифр 5 при первом объединении: ', t)

# Удаляем все вхождения элемента в список
while 5 in a:
    a.remove(5)

# Объединение основного и побочного списка
a.extend(c)

# Подсчет количества троек
t = a.count(3)
print('Кол-во цифр 3 при втором объединении: ', t)

print('Итоговый список: ', a)
```

### Задача №4

В честь своего дня рождения Артём решил закатить вечеринку у себя на даче. Он не стал рассылать приглашения, а просто сообщил всем: «Если хотите — приходите и своих друзей тоже зовите». В ходе вечеринки люди приходили и уходили, ночевать остались не все. К тому же и сама дача не резиновая — на ней помещается всего шесть человек.<br/>
Дан изначальный список гостей — имена тех, кто пришёл к началу:<br/>
guests = ['Петя', 'Ваня', 'Саша', 'Лиза', 'Катя'] <br/>


Напишите программу, которая спрашивает у пользователя, ушёл человек или пришёл новый гость, и, исходя из ответа, добавляет в список или удаляет из него нужное имя. При этом гостей может быть не больше шести. Имена запрашиваются до тех пор, пока пользователь не введёт сообщение «Пора спать».<br/>
Пример:<br/>
Сейчас на вечеринке 5 человек: [‘Петя’, ‘Ваня’, ‘Саша’, ‘Лиза’, ‘Катя’]<br/>
Гость пришёл или ушёл? пришёл<br/>
Имя гостя: Алекс<br/>
Привет, Алекс!<br/>


```python
# Условие (изначальный список гостей)

guests = ['Петя', 'Ваня', 'Саша', 'Лиза', 'Катя']
```

```python
while True:
    print(f'\nСейчас на вечеринке {len(guests)} человек: {guests}') 
    answer = input('Гость пришёл или ушёл? ').strip().lower()
    if answer.strip().lower() != 'пора спать':
        name = input('Имя гостя: ')
        if answer.strip().lower() == 'пришел' and len(guests) < 6:
            guests.append(name)
            print(f'Привет, {name}!')
        elif answer.strip().lower() == 'ушел':
            guests.remove(name)
            print(f'Пока, {name}!')
        else:
            print(f'Прости, {name}, но мест нет.')
    else:
        print('Вечеринка закончилась, все легли спать.')
        break
```

Мы пишем приложение для удобного прослушивания музыки. У Вани есть список из девяти песен группы Depeche Mode. Каждая песня состоит из названия и продолжительности с точностью до долей минут: (см. ниже)<br/>

Из этого списка Ваня хочет выбрать N песен и закинуть их в особый плейлист с другими треками. И при этом ему важно, сколько времени в сумме эти N песен будут звучать.<br/>
Напишите программу, которая запрашивает у пользователя количество песен из списка и затем названия этих песен, а на экран выводит общее время их звучания.<br/>

Пример: <br/>
Сколько песен выбрать? 3 <br/>
Название 1-й песни: Halo <br/>
Название 2-й песни: Enjoy the Silence <br/>
Название 3-й песни: Clean <br/>

Общее время звучания песен: 14,93 минуты 

```python
# Условие

violator_songs = [
    ['World in My Eyes', 4.86],
    ['Sweetest Perfection', 4.43],
    ['Personal Jesus', 4.56],
    ['Halo', 4.9],
    ['Waiting for the Night', 6.07],
    ['Enjoy the Silence', 4.20],
    ['Policy of Truth', 4.76],
    ['Blue Dress', 4.29],
    ['Clean', 5.83]
]
```

### Задача №7

Частная контора даёт в прокат ролики самых разных размеров. Человек может надеть ролики любого размера, которые не меньше размера его ноги. <br/>
Пользователь вводит два списка размеров: N размеров коньков и K размеров ног людей. Реализуйте код, который определяет, какое наибольшее число человек сможет одновременно взять ролики и пойти покататься. 

```python
N = int(input('Кол-во коньков: '))
shoes_size = sorted([int(input(f'Размер {i + 1}-й пары: ')) for i in range(N)])
K = int(input('Кол-во людей: '))
leg_size = sorted([int(input(f'Размер ноги {j + 1}-го человека: ')) for j in range(K)])

match = 0
i, j = 0, 0

while i < N and j < K:
    if shoes_size[i] >= leg_size[j]:  # Если размер коньков подходит
        match += 1                    # Увеличиваем счетчик подходящих пар
        j += 1                        # Переходим к следующему человеку
    i += 1                            # Переходим к следующему размеру коньков

print('\nНаибольшее кол-во людей, которые могут взять ролики: ', match)
```

### Задача №8

N человек, пронумерованных числами от 1 до N, стоят в кругу. Они начинают играть в считалку на выбывание, где каждый K-й по счёту человек выбывает из круга, после чего счёт продолжается со следующего за ним человека.
На вход подаётся количество человек N и номер K. Напишите программу, которая выводит число от 1 до N — это номер человека, который останется в кругу последним.

```python
N = int(input('Кол-во человек: '))
K = int(input('Какое число в считалке? '))
print(f'Значит выбывает каждый {K}-й человек')

lst = list(range(1, N + 1))
index = 0

while len(lst) > 1:
    print('\nТекущий круг людей: ', lst)
    index = (index + K - 1) % len(lst)
    print('Начало счёта с номера', index)
    print('Выбывает человек под номером', lst.pop(index))

print('\nОстался человек под номером', lst[0])
```

### Задача №9

Напишите программу, которая по заданным распискам вычислит, сколько всего должен каждый друг выплатить другим (или получить с других).<br/>
Сначала вводится число N — количество друзей, затем вводится число K — количество долговых расписок, после этого следует K троек чисел: номер друга, взявшего в долг, номер друга, давшего деньги, и сумма. Гарантируется, что ни один друг не брал в долг сам у себя.<br/>
Программа должна вывести «баланс друзей», то есть суммы, которые должны получить или отдать друзья. Положительное число означает, что друг должен получить деньги от других, отрицательное — должен отдать деньги.

```python
N = int(input('Кол-во друзей: '))
K = int(input('Долговых расписок: '))

# Инициализация баланса для каждого друга
balance = [0] * (N + 1)  # Индексы от 1 до N (0 не используется)

for i in range(K):
    print(f'\n{i + 1}-я расписка')
    debtor = int(input('Кому: '))
    creditor = int(input('От кого: '))
    amount = int(input('Сколько: '))
    
    balance[debtor] += amount    # Друг, который получил деньги, увеличивает свой баланс
    balance[creditor] -= amount  # Друг, который дал деньги, уменьшает свой баланс

print('\nБаланс друзей:')
for friend in range(1, N + 1):
    print(f'{friend}: {balance[friend]}')
```

### Задача №10

Последовательность чисел называется симметричной, если она одинаково читается как слева направо, так и справа налево. Например, следующие последовательности являются симметричными: <br/>
1 2 3 4 5 4 3 2 1 <br/>
1 2 1 2 2 1 2 1 <br/>
Пользователь вводит последовательность из N чисел. Напишите программу, которая определяет, какое минимальное количество и каких чисел надо приписать в конец этой последовательности, чтобы она стала симметричной. <br/>

Пример 1:<br/>
Кол-во чисел: 5 <br/>
Число: 1 <br/>
Число: 2 <br/>
Число: 1 <br/>
Число: 2 <br/>
Число: 2 <br/>

Последовательность: [1, 2, 1, 2, 2] <br/>
Нужно приписать чисел: 3 <br/>
Сами числа: [1, 2, 1]

```python
N = int(input('Кол-во чисел: '))
chain = [int(input('Число: ')) for i in range(N)]
print('\nПоследовательность: ', chain)

rev_chain = chain[::-1]

for i in range(N - 1):
    if rev_chain[i] != rev_chain[i + 1]:
        dop_chain = rev_chain[i + 1 : ]
        print('Нужно приписать чисел: ', len(dop_chain))
        print('Сами числа: ', dop_chain)
        break
```
